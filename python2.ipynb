Importando bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import glob
```

- Definindo variáveis com impressão do gráfico
- Solucionando a tarefa
- Solucionando a tarefa bônus


```python
glob.glob ('*txt')
arquivos = glob.glob('dados/*W-TAVG-Trend.txt')
print(arquivos)
```

```python
for nome_arquivo in arquivos:
    
        dados = np.loadtxt(nome_arquivo, comments = '%')
        Anos = dados[:, 0]
        Meses = dados[:, 1]
        ano_decimal = ((Meses - 1)/12 + Anos)
        anomalia_temperatura = dados[:,4]
        
        anomalia_10anos = dados [:, 8]
        incerteza_10anos = dados [:, 9]
        incerteza_95porcento1 = (anomalia_10anos + incerteza_10anos)
        incerteza_95porcento2 = (anomalia_10anos - incerteza_10anos)
    
        plt.figure ()
        plt.plot (Anos, anomalia_10anos, c ='red')
        plt.fill_between (Anos, incerteza_95porcento1, incerteza_95porcento2, color = '#C0C0C0')
        plt.plot (ano_decimal, anomalia_temperatura, c ='#0000FF')
        plt.xlabel ("Anos")
        plt.ylabel ("Temperatura em Cº")
        plt.savefig
    
    
        plt.savefig ( 'gráfico do arquivo ' + nome_arquivo [6: -3] + 'png', format = 'png')
```
